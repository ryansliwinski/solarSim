# imports and constants

```python
import math
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
```

```python
G = 4 * np.pi**2
t = 1
dt = 0.001
N = int(t/dt)
```

# initial conditions

```python
sun_pos = np.array([0.0,0.0])
earth_pos = np.array([1.0, 0.0])
```

# earf

```python
def accel(r):
    r_mag = np.linalg.norm(r)
    return -(G * r )/ r_mag**3
```

```python
a0 = accel(earth_pos)
r_mag = np.linalg.norm(earth_pos)
v_mag = np.sqrt(np.linalg.norm(a0) * r_mag)
print(v_mag)
```

```python
pos = []
a_old = accel(earth_pos)
r = np.array([-earth_pos[-1],earth_pos[0]]) / r_mag

earth_v = v_mag * r

for i in range(N):
    earth_v += a_old * (dt/2)
    earth_pos += earth_v * dt
    a_new = accel(earth_pos.copy())
    earth_v += a_new * (dt/2)
    a_old = a_new
    pos.append(earth_pos.copy())
    #print("x=",earth_pos)
    #print("v=",earth_v)
    #print("a=",a_old)
    
#np.dot(earth_v, earth_pos)
```

# plot

```python
x,y = zip(*pos)
fig, ax = plt.subplots()
ax.scatter(0,0,s=200,color='gold')
ax.plot(x,y)
plt.show()
```
